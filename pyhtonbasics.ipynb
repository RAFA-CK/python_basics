What is the use of a function?

A function is used to organize your code into reusable, manageable pieces that perform specific tasks.

Instead of writing the same code again and again, you can put it inside a function and just call that function whenever you need it.

```python
from re import A
def greet(name):
  print("hello " + name + "!")

greet("alice")

def add(a,b):
  print(a+b)
add(3,4)

def multiple(a,b,c):
  print(a*b*c)
multiple(2,3,4)
```
